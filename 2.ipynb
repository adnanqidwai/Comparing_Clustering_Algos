```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
compact=pd.read_csv('Compact.csv')
skewed=pd.read_csv('Skewed.csv')
subClusters=pd.read_csv('Subclusters.csv')
wellSeperated=pd.read_csv('WellSeperated.csv')
```

```python
compact.plot(x='x1', y='x2', kind='scatter')
```

```python
skewed.plot(x='x1', y='x2', kind='scatter')
```

```python
subClusters.plot(x='x1', y='x2', kind='scatter')
```

```python
wellSeperated.plot(x='0', y='1', kind='scatter')
```

```python
n_clusters=5
```

```python
Kmeans_compact=KMeans(n_clusters=n_clusters, random_state=0).fit(compact)
Kmeans_skewed=KMeans(n_clusters=n_clusters, random_state=0).fit(skewed)
Kmeans_subClusters=KMeans(n_clusters=n_clusters,random_state=0).fit(subClusters)
Kmeans_wellSeperated=KMeans(n_clusters=n_clusters, random_state=0).fit(wellSeperated)
```

```python
plt.scatter(skewed['x1'], skewed['x2'], c=Kmeans_skewed.labels_, cmap='rainbow')
```

```python
plt.scatter(subClusters['x1'], subClusters['x2'], c=Kmeans_subClusters.labels_, cmap='rainbow')
```

```python
plt.scatter(wellSeperated['0'], wellSeperated['1'], c=Kmeans_wellSeperated.labels_, cmap='rainbow')
```

```python
DBSCAN_compact=DBSCAN(eps=2, min_samples=3).fit(compact)
plt.scatter(compact['x1'], compact['x2'], c=DBSCAN_compact.labels_, cmap='rainbow')
```

```python
DBSCAN_skewed=DBSCAN(eps=2, min_samples=1).fit(skewed)
plt.scatter(skewed['x1'], skewed['x2'], c=DBSCAN_skewed.labels_, cmap='rainbow')
```

```python
DBSCAN_subClusters=DBSCAN(eps=3.5, min_samples=40).fit(subClusters)
plt.scatter(subClusters['x1'], subClusters['x2'], c=DBSCAN_subClusters.labels_, cmap='rainbow')
```

```python
DBSCAN_wellSeperated=DBSCAN(eps=30, min_samples=3).fit(wellSeperated)
plt.scatter(wellSeperated['0'], wellSeperated['1'], c=DBSCAN_wellSeperated.labels_, cmap='rainbow')
```
